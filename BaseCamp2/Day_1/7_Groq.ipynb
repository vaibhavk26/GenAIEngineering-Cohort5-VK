### Groq Usage for LLM ###
**Library Need**  
To use the APIs provided by the groq platform, the library 'groq' is required.  
Ensure to have performed 'pip install groq dotenv'  
The API key from your Groq account would be required to execute this code


```python
# Import required class from Groq
from groq import Groq
from dotenv import load_dotenv

# Initialise an client object with API key
load_dotenv ()
Client = Groq ()

# Client = Groq (api_key='Your API Key')
```

**Invoke for chat Completion**  
>The APIs can be used for invoking LLM models available in Groq for chat completion  
>Try with differet models and different prompt messages

```python
#prompt = "ML Vs GenAI ..?"

prompt ='''{
  "purpose": "Gen-AI career preference prompt",
  "instructions": "You are a career-counsellor assistant helping users discover Gen-AI career paths.\nWorkflow:\n• Step 1 – Reveal each statement below one by one. After each, wait for the user to reply with -1, 0, or 1.\n• Step 2 – When all 10 ratings are collected, sum scores per category. Higher totals mean stronger interest. Briefly list the top areas and ask: \"Does this reflect your interests? (yes/no)\".\n• Step 3 – If the user says yes, suggest 3–5 fitting career titles. For each give: main Gen-AI area, why it fits, and two practical starting steps.\nUse short, crisp sentences and bullet points.",
  "statements": [
    { "id": 1, "category": "Data", "text": "I enjoy exploring large datasets to find meaningful patterns." },
    { "id": 2, "category": "Data Analysis", "text": "I like using statistical tools to turn raw data into insights." },
    { "id": 3, "category": "ML Ops", "text": "Automating model deployment and monitoring excites me." },
    { "id": 4, "category": "Building Applications", "text": "Building end-to-end software products motivates me." },
    { "id": 5, "category": "Agents", "text": "Designing autonomous AI agents that act without constant oversight appeals to me." },
    { "id": 6, "category": "Chatbots", "text": "Crafting chatbots that hold natural conversations is rewarding." },
    { "id": 7, "category": "Evals & Testing", "text": "I enjoy stress-testing AI models to measure real-world performance." },
    { "id": 8, "category": "Cost Control & Reduction", "text": "Finding ways to cut compute costs in ML workflows motivates me." },
    { "id": 9, "category": "Fine-Tuning", "text": "Adapting pre-trained models to niche use cases interests me." },
    { "id": 10, "category": "Guardrails", "text": "Implementing robust safety and ethical guardrails for AI systems matters to me." }
  ],
  
}'''

# Prompt needs to be organised as message (list of dict)
messages=[
    {
        "role": "user",
        "content": prompt,
    }
]
completion = Client.chat.completions.create(
    messages=messages,    
    # model="llama-3.3-70b-versatile",
    model="openai/gpt-oss-120b",
    # model="llama-3.1-8b-instant",
    stop=None,
)

print (completion.choices[0].message.content)
```
